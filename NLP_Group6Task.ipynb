```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import re
import operator
d=dict()
natgeo=open("natgeo.txt","r",encoding="utf8")
compare=open("Compare.txt","r",encoding="utf8")
compareContent=compare.read()
compareContent=compareContent.lower()
natGeoContent=natgeo.read()
natGeoContent=natGeoContent.lower()

#remove punctuations
natGeoContent = re.sub(r'[.,"\'-?:!;]', '', natGeoContent)
compareContent = re.sub(r'[.,"\'-?:!;]', '',compareContent)

#number of words
words=natGeoContent.split()
compareWords=compareContent.split()
print("no of words in document are-")
print(int(len(words))+1)
print("word list and frequency in ai-")
for word in words:
    if word in d:
        d[word]=d[word]+1
    else:
        d[word]=1
for key in list(d.keys()):
    print(key, ":",d[key])

#now removing stop words
stop_words=nltk.corpus.stopwords.words('english')
stop_words.append('.')
stop_words.append(',')
stop_words.append('(')
stop_words.append(')')
d2=dict()
count=0;
filtered_words= [w for w in words if not w in stop_words]

for word in filtered_words: 
    if word in d2:
        d2[word]=d2[word]+1
    else:
        count=count+1
        d2[word]=1
        
#now arranging in descending order by term count
sorted_d = dict(sorted(d2.items(), key=operator.itemgetter(1),reverse=True))
print("frequency of words are-")
for key in list(sorted_d.keys()):
    print(key, ":",sorted_d[key])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatized_words=[]
lemmatizer=WordNetLemmatizer()
for word in words:
    lemmatized_words.append(lemmatizer.lemmatize(word))
```
